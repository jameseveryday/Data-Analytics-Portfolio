# crimes.csv

| Column     | Description              |
|------------|--------------------------|
| `'DR_NO'` | Division of Records Number: Official file number made up of a 2-digit year, area ID, and 5 digits. |
| `'Date Rptd'` | Date reported - MM/DD/YYYY. |
| `'DATE OCC'` | Date of occurrence - MM/DD/YYYY. |
| `'TIME OCC'` | In 24-hour military time. |
| `'AREA NAME'` | The 21 Geographic Areas or Patrol Divisions are also given a name designation that references a landmark or the surrounding community that it is responsible for. For example, the 77th Street Division is located at the intersection of South Broadway and 77th Street, serving neighborhoods in South Los Angeles. |
| `'Crm Cd Desc'` | Indicates the crime committed. |
| `'Vict Age'` | Victim's age in years. |
| `'Vict Sex'` | Victim's sex: `F`: Female, `M`: Male, `X`: Unknown. |
| `'Vict Descent'` | Victim's descent:<ul><li>`A` - Other Asian</li><li>`B` - Black</li><li>`C` - Chinese</li><li>`D` - Cambodian</li><li>`F` - Filipino</li><li>`G` - Guamanian</li><li>`H` - Hispanic/Latin/Mexican</li><li>`I` - American Indian/Alaskan Native</li><li>`J` - Japanese</li><li>`K` - Korean</li><li>`L` - Laotian</li><li>`O` - Other</li><li>`P` - Pacific Islander</li><li>`S` - Samoan</li><li>`U` - Hawaiian</li><li>`V` - Vietnamese</li><li>`W` - White</li><li>`X` - Unknown</li><li>`Z` - Asian Indian</li> |
| `'Weapon Desc'` | Description of the weapon used (if applicable). |
| `'Status Desc'` | Crime status. |
| `'LOCATION'` | Street address of the crime. |

```python
# Re-run this cell
#Â Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
crimes = pd.read_csv("crimes.csv", parse_dates=["Date Rptd", "DATE OCC"], dtype={"TIME OCC": str})
crimes.head()
```

```python
# Start coding here
# Use as many cells as you need
```

```python
# Hour that has the highest frequency of crimes

# Extract the hour from the 'TIME OCC' column
crimes['Hour'] = crimes["TIME OCC"].str[:2].astype(int)

# Find the hour with the highest frequency of crimes
peak_crime_hour = crimes['Hour'].mode().iloc[0]
print("Hour with the highest frequency of crimes:", peak_crime_hour)

# Plot
plt.figure(figsize=(10, 6))
sns.countplot(data=crimes, x='Hour', color='green')
plt.title('Frequency of Crimes per Hour')
plt.xlabel('Hour of Day, 24-hour format')
plt.ylabel('Frequency')
plt.show()
```

```python
# Area that has the largest frequency of night crimes
# Crimes committed between 10pm and 3:59am

# Filter the dataset for night crimes
night_crimes = crimes[(crimes['Hour'] >= 22) | (crimes['Hour'] < 4)]

# Find the area with the largest frequency of night crimes
peak_night_crime_location = night_crimes['AREA NAME'].mode().iloc[0]

print("Area with the largest frequency of night crimes:", peak_night_crime_location)

# Plot
plt.figure(figsize=(10, 6))
night_crimes['AREA NAME'].value_counts().nlargest(10).plot(kind='bar', color='Orange')
plt.title('Top 10 Areas with the Largest Frequency of Night Crimes')
plt.xlabel('Area Name')
plt.ylabel('Frequency')
plt.xticks(rotation=12.5)
plt.show()
```

```python
# Number of crimes committed against victims by age group
# Age Groups: 0-17, 18-25, 26-34, 35-44, 45-54, 55-65, 65+

# Define age bins and labels
age_bins = [0, 17, 25, 34, 44, 54, 64, np.inf]
age_labels = ["0-17", "18-25", "26-34", "35-44", "45-54", "55-64", "65+"]

# Add a new column using pd.cut() to bin values into discrete intervals
crimes["Age Group"] = pd.cut(crimes["Vict Age"],
                               bins=age_bins,
                               labels=age_labels)

#Â Find the category with the largest frequency
victim_ages = crimes["Age Group"].value_counts()
print(victim_ages)

# Plot
plt.figure(figsize=(10, 6))
sns.countplot(data=crimes, x='Age Group')
plt.title('Number of crimes committed by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Frequency')
plt.show()
```
